```python
def update_hidden_state(X, W_xh, H, W_hh):
    H_new = tf.tanh(tf.matmul(X, W_xh) + tf.matmul(H, W_hh))
    return H_new
H_t = update_hidden_state(X,W_xh,H, W_hh)
H_t
```
