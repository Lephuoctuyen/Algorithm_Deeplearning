Việc tham đôi gamma, beta, phương sai và giá trị kỳ vọng ví bài toán ví dụ cơ bản để trực quan nó:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Giả lập dữ liệu
np.random.seed(42)
data = np.random.randn(1000, 1) * 2 + 3  # Tạo dữ liệu có mean=3 và std=2

# Normalization (chuẩn hóa) ban đầu
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
normalized_data = (data - mean) / std

# Khởi tạo gamma và beta
gamma_values = [0.5, 1.0, 2.0]
beta_values = [-1.0, 0.0, 1.0]

# Tạo plot với các subplots
fig, axes = plt.subplots(len(gamma_values), len(beta_values), figsize=(12, 8))

for i, gamma in enumerate(gamma_values):
    for j, beta in enumerate(beta_values):
        # Chuẩn hóa với gamma và beta hiện tại
        bn_data = gamma * normalized_data + beta

        # Tính mean và variance
        mean_bn = np.mean(bn_data)
        var_bn = np.var(bn_data)

        # Plot PDF
        x = np.linspace(bn_data.min()-1, bn_data.max()+1, 100)
        pdf = norm.pdf(x, mean_bn, np.sqrt(var_bn))

        axes[i, j].plot(x, pdf, color='blue', lw=2)
        axes[i, j].fill_between(x, pdf, color='blue', alpha=0.2)

        # Hiển thị mean và variance
        axes[i, j].axvline(mean_bn, color='red', linestyle='--', label=f'Mean: {mean_bn:.2f}')
        axes[i, j].text(mean_bn, max(pdf), f'Mean: {mean_bn:.2f}', color='red', ha='right')
        axes[i, j].text(mean_bn, max(pdf) * 0.8, f'Var: {var_bn:.2f}', color='green', ha='right')

        # Set titles
        axes[i, j].set_title(f'γ={gamma}, β={beta}')
        axes[i, j].legend()

plt.tight_layout()
plt.show()
```

Lưu ý rằng trong ngữ cảnh của các lớp tích chập, chuẩn hóa theo lô (batch normalization) **vẫn được định nghĩa tốt ngay cả khi đối với các minibatch có kích thước là 1**: suy cho cùng, chúng ta có tất cả các vị trí trên một hình ảnh để tính trung bình. Do đó, giá trị trung bình và phương sai được xác định rõ ràng, ngay cả khi nó chỉ nằm trong một quan sát duy nhất. Sự cân nhắc này đã dẫn đến việc Ba và cộng sự (2016) giới thiệu khái niệm **layer normalization (chuẩn hóa theo lớp)**. Nó **hoạt động giống như chuẩn hóa theo lô, chỉ khác ở chỗ nó được áp dụng cho từng quan sát một lần**. Do đó, cả hệ số dịch chuyển và hệ số tỉ lệ đều là các số vô hướng. Đối với một vector \( \mathbf{x} \) có \( n \) chiều, chuẩn hóa theo lớp được cho bởi:

$$
\mathbf{x} \rightarrow \text{LN}(\mathbf{x}) = \frac{\mathbf{x} - \hat{\mu}}{\hat{\sigma}},
$$

trong đó việc chuẩn hóa và dịch chuyển được áp dụng theo từng hệ số và được cho bởi:

$$
\hat{\mu} = \frac{1}{n} \sum_{i=1}^{n} x_i \quad \text{và} \quad \hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{\mu})^2 + \epsilon.
$$

Như trước đây, chúng ta thêm một lượng nhỏ \epsilon > 0  để ngăn ngừa việc chia cho 0. Một trong những lợi ích chính của việc sử dụng chuẩn hóa theo lớp là nó ngăn ngừa hiện tượng phân kỳ (divergence). Suy cho cùng, bỏ qua \( \epsilon \), đầu ra của chuẩn hóa theo lớp không phụ thuộc vào tỉ lệ. Cụ thể, chúng ta có:

$$
\text{LN}(\mathbf{x}) \approx \text{LN}(\alpha \mathbf{x})
$$

với bất kỳ giá trị nào của \( \alpha \neq 0 \). Điều này trở thành một đẳng thức khi \( |\alpha| \rightarrow \infty \) (tính gần đúng là do sự hiện diện của \( \epsilon \) trong phương sai).
